```python
import pandas as pd
import numpy as np
```

```python
url = 'https://www.data.gouv.fr/fr/datasets/r/5c4e1452-3850-4b59-b11c-3dd51d7fb8b5'
df = pd.read_csv(url, dtype={'dep': 'str'})

df.head()
```

```python
df['reg'].drop_duplicates().values
```

```python
df['dep'].drop_duplicates().values
```

```python
# Check types
df.dtypes
```

```python
# Set multi index
df = df.set_index(['dep', 'date']).sort_index()
```

```python
df.tail(10)
```

```python
# Get info
print(df.info())
```

```python
# Look into R
df[df['R'].isnull() == False].loc[('75', slice(None)),'R'].plot()
```

```python
# Look into both pos'
df[df['pos'].isnull() == False].loc[('75', slice(None)),'pos':'pos_7j'].tail()
```

```python
df = df.drop(['R', 'cv_dose1', 'rad', 'incid_rad'], axis=1)
```

```python
# Set tx_pos NaN to mean
x = df.reset_index()['tx_pos'].mean()
df['tx_pos'].fillna(x, inplace=True)
```

```python
# Set tx_incid NaN to mean
x = df.reset_index()['tx_incid'].mean()
df['tx_incid'].fillna(x, inplace=True)
```

```python
df.info()
```

```python
df['pos'].fillna(0, inplace=True)
df['pos_7j'].fillna(0, inplace=True)
```

```python

columns = ['incid_hosp', 'incid_rea', 'incid_dchosp', 'reg_incid_rea']
df[columns] = df[columns].fillna(0)
```

```python
# Corr matrix
df.corr()
```
